```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
scaler.fit(data_set.drop('diagnosis_M',axis=1))
```

```python
scaled_features=scaler.transform(data_set.drop('diagnosis_M',axis=1))
```

```python
df_feat=pd.DataFrame(scaled_features,columns=data_set.columns[:-1])
```

```python
from sklearn.model_selection  import train_test_split
```

```python
X=df_feat
y=final_dataset[['diagnosis_M']]
train_test_split= X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions=dtree.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rforest=RandomForestClassifier()
```

```python
rforest.fit(X_test,y_test)
```

```python
predictions=rforest.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
